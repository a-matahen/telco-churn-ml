```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
os.makedirs("../outputs", exist_ok=True)
sns.set_style("darkgrid")
df =pd.read_csv("../data/telco_churn_clean.csv")
```

```python
plt.figure(32,figsize=(7,5),dpi=200)
df['Churn'].value_counts().plot(kind='bar',color=["#3B8BD4", "#D85A30"])
plt.title("Customer churn distribution ")
plt.xlabel("Charn")
plt.ylabel("number of customer")
plt.xticks(ticks=[0,1],labels=["stayed","left"],rotation=0)
plt.tight_layout()
plt.savefig("../outputs/02_churn.png")
plt.close()
```

```python
plt.figure(42,figsize=(7,5),dpi=200)
churn_by_contract=df.groupby("Contract")["Churn"].mean().sort_values()
churn_by_contract.plot(kind="barh",color="#1D9E75")
plt.title("Churn rate by contract type ")
plt.xlabel("Churn rate")
plt.tight_layout()
plt.savefig("../outputs/03_churn.png")
plt.close()
print(churn_by_contract)
```

```python
plt.figure(44,figsize=(7,5),dpi=200)
sns.histplot(data=df, x="tenure", hue="Churn", multiple="stack", bins=30, palette=["#3B8BD4", "#D85A30"])
plt.title("Customer tenure (months) vs churn")
plt.xlabel("Tenure (months)")
plt.tight_layout()
plt.savefig("../outputs/05_tenure_vs_churn.png")
plt.close()

```

```python
plt.figure(45,figsize=(6, 4),dpi=200)
sns.boxplot(data=df, x="Churn", y="MonthlyCharges",hue="Churn", palette=["#3B8BD4", "#D85A30"])
plt.title("Monthly charges by churn status")
plt.tight_layout()
plt.savefig("../outputs/05_charges_vs_churn.png")
plt.close()
```
